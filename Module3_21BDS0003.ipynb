# Time Series

```python
# Assuming the dataset is already loaded into df
n_rows = attr.shape[0]

# Simulate a 'HireDate' column by creating a sequence of dates
attr['HireDate'] = pd.date_range(start='2010-01-01', periods=n_rows, freq='M')

# Set 'HireDate' as the index for time series analysis
attr.set_index('HireDate', inplace=True)

print(attr.head())
```

```python
# Get the start and end dates
start_date = attr.index.min()
end_date = attr.index.max()

print(f"Start Date: {start_date}")
print(f"End Date: {end_date}")
```

```python
# Check frequency
frequency = pd.infer_freq(attr.index)
print(f"Dataset frequency: {frequency}")
```

```python
print(attr.describe())
```

```python
from statsmodels.tsa.seasonal import seasonal_decompose

decomposition = seasonal_decompose(attr['MonthlyIncome'], model='additive')
decomposition.plot()
```

# Plotting Time Series for 'MonthlyIncome'

```python
plt.figure(figsize=(10, 6))
attr['MonthlyIncome'].plot()
plt.title('Time Series of Monthly Income')
plt.xlabel('Hire Date')
plt.ylabel('Monthly Income')
plt.show()
```

```python
attr.plot()
plt.title('Employee Attrition Dataset')
plt.legend(loc='upper left', bbox_to_anchor=(1, 1))

plt.show()
```

# Creating a time index

```python
time_idx = np.arange(len(attr))
slope, intercept = np.polyfit(time_idx, attr['MonthlyIncome'], 1)

plt.plot(attr.index, attr['MonthlyIncome'], label='Monthly Income')

plt.plot(attr.index, intercept + slope*time_idx, 'r', label='Regression Line')
plt.legend()
plt.show()
```

```python
attr['cycle'] = attr.index.month
print(attr['cycle'])
```

```python
log_data = np.log(attr['MonthlyIncome'])

plt.plot(log_data)
plt.title('Log Transformed Monthly Income Data')
plt.show()
```

# Apply logarithm transformation

```python
attr['log_MonthlyIncome'] = np.log(attr['MonthlyIncome'])

attr['stationary_log_income'] = attr['log_MonthlyIncome'].diff()
attr['stationary_log_income'].dropna().plot(title="Stationary Log of Monthly Income")
```
