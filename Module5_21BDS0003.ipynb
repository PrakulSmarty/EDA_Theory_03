# Visualizing dendrograms with different distance methods

```python
def compute_dendrogram(data, distance_method, linkage_method='ward'):
    distance_matrix = pdist(data, metric=distance_method)
    print(f"\nDistance Matrix (Method: {distance_method})")

    matrix_preview = squareform(distance_matrix)[:10, :10]
    print(matrix_preview)

    hc = linkage(distance_matrix, method=linkage_method)

    # Plot the dendrogram
    plt.figure(figsize=(10, 6))
    plt.title(f"Dendrogram using {distance_method} distance (21BDS0003)")
    plt.xlabel("Observations")
    plt.ylabel("Height")
    dendrogram(hc)
    plt.show()
```

# Euclidean Distance

```python
compute_dendrogram(dataset.values, "euclidean")
```

# Manhattan Distance

```python
compute_dendrogram(dataset.values, "cityblock")
```

# Maximum Distance

```python
compute_dendrogram(dataset.values, "chebyshev")
```

# Canberra Distance

```python
compute_dendrogram(dataset.values, "canberra")
```

# Binary Distance

```python
compute_dendrogram(dataset.values, "jaccard")
```

# Minkowski Distance

```python
compute_dendrogram(dataset.values, "minkowski")
```

# K-Means Clustering and Visualization

```python
def plot_kmeans_clusters_attrition(data, n_clusters):
    kmeans = KMeans(n_clusters=n_clusters, init='k-means++', random_state=42)
    labels = kmeans.fit_predict(data)

    plt.figure(figsize=(8, 6))
    plt.scatter(data[:, 0], data[:, 1], c=labels, cmap='plasma', s=50)
    plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:, 1], s=200, c='red', label='Centroids')
    plt.title('K-Means Clustering - 21BDS0003')
    plt.xlabel('Feature 1 (Scaled MonthlyIncome)')
    plt.ylabel('Feature 2 (Scaled YearsAtCompany)')
    plt.legend()
    plt.show()

    silhouette_avg = silhouette_score(data, labels)
    print(f'Silhouette Score 21BDS0003: {silhouette_avg}')

plot_kmeans_clusters_attrition(attrition_scaled, attrition_clusters)
```
